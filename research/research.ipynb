```python
import pandas as pd
```

```python
data = pd.read_csv(r'C:\Users\ayush\MLOPS\LOAN-DETECTION\data\archive (33)\loan_data.csv')
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, drop='first')
```

```python
encoded_data = encoder.fit_transform(data)
```

```python
pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(data.columns))
```

```python
data.head(n=5)
```

```python
data = pd.get_dummies(data, columns=['person_gender', 'person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file'], drop_first=True)
```

```python
data
```

```python
import numpy as np
```

```python
for col in data:
    if data[col].dtype == 'bool':
        data[col] = np.where(data[col] == False, 0 , 1)
```

```python
data
```

```python
data
```

```python
import tensorflow
```

```python
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Sequential
```

```python

l1 = Dense(units=256, activation='relu' , input_dim = 22)
l2 = Dense(units=128 , activation='relu')
l3 = Dense(units=64 , activation='relu')
l4 = Dense(units=32 , activation='relu')
l5 = Dense(units=1, activation='sigmoid')
```

```python
model = Sequential()
```

```python
model.add(l1)
model.add(l2)
model.add(l3)
model.add(l4)
model.add(l5)
```

```python
model.compile(optimizer='adam' , metrics=['accuracy'], loss='binary_crossentropy')
```

```python
x = data.drop(columns = ['loan_status'])
y = data['loan_status']
```

```python
x
```

```python
y.shape
```

```python
model.fit(x , y , epochs=10)
```
